```python
import sys
from config import *
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
from time import sleep
from random import randint
```

```python
### We import the top_100 songs 

top_100 = pd.read_csv('top_100.csv')
top_100.head(10)
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= client_id,
                                                           client_secret= client_secret))
```

```python
### In this function, you input a list of songs and a list of their respective artist, and it
### returns the song id

def search_song(songs, artists):
    id_song = []
    for song, artist in zip(songs, artists):
        try:
            results = sp.search(q="track:"+song+" artist:"+artist, limit=1)
            ids = results['tracks']['items'][0]["id"]
            id_song.append(ids)
        except:
            ids = ""
            id_song.append(ids)
    return id_song
```

```python
### Here I am generating the list of songs and the list of artists

songs1 = list(top_100["song"])
print(songs1)
artists1 = list(top_100["artist"])
print(artists1)
```

```python
### Here I am running the function with my two list, and generating a list of ids

id_song = search_song(songs1, artists1)
print(id_song)
```

```python
id_song = pd.DataFrame({"song_id":id_song})
id_song
```

```python
### Here I am concatenating the top_100 df, with the corresponding ids

top_100 = pd.concat([top_100, id_song], axis = 1)
top_100.head(10)
```

```python
### Here I am dropping the rows with no ids

top_100 = top_100.set_index("song_id").drop(index = "", axis = 0).reset_index()
top_100
```

```python
### This function retrieves the audio features from a list of song ids 

def get_audio_features(list_song_id):
    data = pd.DataFrame(columns=['danceability', 'energy', 'key', "loudness", "mode",
                          "speechiness", "acousticness", "instrumentalness", "liveness",
                          "valence", "tempo", "type", "id", "uri", "track_href", "analysis_url",
                          "duration_ms", "time_signature"])
    for i in list_song_id:
        try:
            my_dict = sp.audio_features([i])[0]
            df = pd.DataFrame.from_dict(my_dict, orient = "index").T
            data = pd.concat([data, df], axis = 0).reset_index(drop = True)
        except:
            pass
    return data 
```

```python
### Here I am implementing the function with the song ids from the top_100 (actually top 77)

ids = list(top_100["song_id"])
features = get_audio_features(ids)
features
```

```python
### This function concatenates two dataframes

def add_audio_features(df, audio_features_df):
    data = pd.concat([df, audio_features_df], axis = 1)
    return data
```

```python
### Here I am implementing the function

top_100_features = add_audio_features(top_100, features)
top_100_features = top_100_features.drop(["song_id"], axis = 1)
top_100_features
```

```python
### Here I will do all the steps above but with the random_songs
### Additionally, I will divide the random_songs in chunks of 50 as to avoid overloading getting
### blocked from the spotify api

random_songs = pd.read_csv('random_songs.csv')
random_songs.head(10)
```

```python
songs2 = list(random_songs["song"])
artists2 = list(random_songs["artist"])
```

```python
id_song2 = search_song(songs2, artists2)
print(id_song2)
```

```python
id_song2 = pd.DataFrame({"song_id":id_song2})
id_song2
```

```python
random_songs = pd.concat([random_songs, id_song2], axis = 1)
random_songs.head(10)
```

```python
random_songs = random_songs.set_index("song_id").drop(index = "", axis = 0).reset_index()
random_songs
```

```python
### Here I am generating the different chunks of the song_id list

chunksize = 50
list_chunked = [random_songs[i:i + chunksize] for i in range(0, len(random_songs), chunksize)]
```

```python
data2 = pd.DataFrame(columns=['danceability', 'energy', 'key', "loudness", "mode",
                          "speechiness", "acousticness", "instrumentalness", "liveness",
                          "valence", "tempo", "type", "id", "uri", "track_href", "analysis_url",
                          "duration_ms", "time_signature"])

for i in range(0,len(list_chunked)):
    print("Getting audio features for list_chunked number", i)
    df = get_audio_features(list(list_chunked[i]["song_id"]))
    data2 = pd.concat([data2, df], axis = 0).reset_index(drop = True)
    sleep(30)

display(data2)
```

```python
random_songs_features = add_audio_features(random_songs, data2)
random_songs_features = random_songs_features.drop(["song_id"], axis = 1)
print(random_songs_features.shape)
random_songs_features
```
